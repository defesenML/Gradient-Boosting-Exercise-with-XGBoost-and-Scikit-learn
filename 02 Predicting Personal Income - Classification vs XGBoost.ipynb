```python
# Column names can be renamed as follows:
df_census.columns=['age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status', 'occupation', 'relationship', 'race', 'sex', 'capital-gain', 'capital-loss', 'hours-per-week', 'native-country', 'income']

df_census.head()
```

#### non-numerical columns
All columns with "object" type need to be transformed into numerical columns.
Pandas 'get_dummies' method takes the non-numerical unique values of every column and converts them into their own column, with 1 indicating presence and 0 indicating absence. 

```python
# we can eliminate "education" column since it's a duplicate of education_num:
df_census = df_census.drop(['education'], axis=1)
```

```python
# use pd.get_dummies to transform the non-numerical columns into numerical columns:
df_census = pd.get_dummies(df_census)

df_census.head()
```

#### Predictor and target columns

```python
#The target column is whether or not someone makes 50K. Two columns, df_census['income_<=50K'] and df_census['income_>50K'], are used to determine whether someone makes 50K. 
#Since either column will work, we delete df_census['income_ <=50K']:
```

```python
df_census = df_census.drop('income_ <=50K', axis=1)
```

```python
X = df_census.iloc[:,:-1]
y = df_census.iloc[:,-1]
```

```python
##### Saving data for future use:
df_census.to_csv('census_cleaned.csv', index=False)
```

### Logistic Regression

```python
# import logistic regression:
from sklearn.linear_model import LogisticRegression
```

#### The cross-validation function

```python
from sklearn.metrics import mean_squared_error

import numpy as np
```

```python
from sklearn.model_selection import cross_val_score
```

```python
def cross_val(classifier, num_splits=10):    
    model = classifier     
    scores = cross_val_score(model, X, y, cv=num_splits)    
    print('Accuracy:', np.round(scores, 2))    
    print('Accuracy mean: %0.2f' % (scores.mean()))
```

```python
cross_val(LogisticRegression())
```

### The XGBoost classifier

```python
#import classifier of XGBoost:
from xgboost import XGBClassifier
```

```python
# Since there are 94 columns, and XGBoost is an ensemble method, meaning that it combines many models for each run, each of which includes 10 splits, we are going to limit n_estimators, the number of models, to 5.
cross_val(XGBClassifier(n_estimators=5))

#Accuracy: [0.85 0.86 0.87 0.85 0.86 0.86 0.86 0.87 0.86 0.86]
#Accuracy mean: 0.86

# XGBoost is the winner again..
```
